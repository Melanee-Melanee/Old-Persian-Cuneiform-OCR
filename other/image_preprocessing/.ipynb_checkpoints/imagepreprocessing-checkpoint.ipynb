```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image_path= '/content/184144001.jpg'
```

```python
image = cv2.imread(image_path)
```

```python
cv2_imshow(image)
```

```python
# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
  # Apply a binary threshold to get a black and white image
_, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
```

```python
# Remove noise using morphological operations
# Create a kernel
kernel = np.ones((3, 3), np.uint8)
```

```python
# Dilation followed by erosion (closing) to remove noise
cleaned_image = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel)
```

```python
# Invert the image so the background is white and the inscriptions are black
final_image = cv2.bitwise_not(cleaned_image)
```

```python
cv2_imshow(final_image)
```

```python
inverted_image = cv2.bitwise_not(image)
```

```python
cv2_imshow(inverted_image)
```
